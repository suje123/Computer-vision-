### how to use camera by the help of Cv2

```python
cap = cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    cv2.imshow("Camera",frame)
    if cv2.waitKey(1) ==13:
        break
cap.release()
cv2.destroyAllWindows()
```

### creat shapes using OpenCV and matplotlib

```python
img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
cv2.rectangle(img,pt1=(384,10),pt2=(500,150),color=(0,255,0),thickness=9)
plt.imshow(img)
```

```python
cv2.rectangle(img,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=9)
plt.imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=50,color=(200,210,100),thickness=10)
plt.imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=50,color=(200,210,100),thickness=-1)
plt.imshow(img)
```

```python
cv2.line(img,pt1=(0,0),pt2=(511,511),color=(180,50,60),thickness=6)
plt.imshow(img)
```
