### blending technic

```python
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
```

```python
img1.shape
```

```python
img2.shape
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
watermark = cv2.addWeighted(src1 = img1, alpha=0.8,src2=img2,beta=0.4,gamma=0.5)
```

```python
plt.imshow(watermark)
```

### overlay

```python
img1 = cv2.imread('sujeet.JPG')
img2 = cv2.imread('tsujeet.jpg')
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.resize(img2, (1500,1500))
```

```python
img2.shape
```

```python
large = img1
small = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
large[y_offset:y_offset+small.shape[0],x_offset:x_offset+small.shape[1]]=small
```

```python
plt.imshow(large)
```
