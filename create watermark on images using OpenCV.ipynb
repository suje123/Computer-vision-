```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('sujeet.JPG')
img2 = cv2.imread('nature.jpg')
```

```python
img1.shape
```

```python
img2.shape
```

```python
plt.imshow(img1)
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
## i have to make both image same shape
img1 = cv2.resize(img1,(1350,1350))
img2 = cv2.resize(img2,(1350,1350))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
watermark = cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.7,gamma=0)
```

```python
plt.imshow(watermark)
```
