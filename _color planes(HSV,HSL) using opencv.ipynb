```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sujeet.JPG')
```

```python
plt.imshow(img) # opencv BGR, matplotlib RGB
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img2)
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img2)
```
